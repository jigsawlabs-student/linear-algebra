### Introduction

Now there are two main operations that we perform in linear algebra, the first is adding.  Let's start learn some vector addition.

### Back to Linear Equations

Let's say that we have the following data:

|ad spending    | 
| ------------- |
|    800        |
|    1500        |
|    2000      | 
|    3500      | 
|    4000      | 

Now remember that we believe that ad spending multiplied by some number equals the number of T-shirts sold.  So when we try to predict the number of T-shirts sold we write it as the following:

$$800x $$
$$1500x $$
$$2000x$$
$$3500x  $$
$$4000x $$

Here the numbers 800 to 400 are our feature variables for ad spending, and $x$ is the related coefficient.

Let's say that $x = 2$.  That is for every dollar on ad spending sold we then sell two t-shirts.

$$800x = 1600$$
$$1500x = 3000$$
$$2000x = 4000 $$
$$3500x = 7500 $$
$$4000x = 4000 $$

When $x = 2$.

### What's the problem?

Now to perform calculate the above, we had to perform five separate operations.  Even though, when first articulated what we were doing, we described it more abstractly: multiply our set of numbers by two.  Vectors allows us to stay in this higher level of abstraction.  We simply treat each of our ad spend feature varibles as a vector, and then we can rewrite the above like so.

$2* \begin{pmatrix}
    800 \\
    1500 \\
    2000 \\
    3500 \\
    4000 \\
\end{pmatrix} = \begin{pmatrix}
    2* 800 \\
    2*1500 \\
    2*2000 \\
    2*3500 \\
    2*4000 \\
\end{pmatrix} = \begin{pmatrix}
    1600 \\
    3000 \\
    4000 \\
    7000 \\
    8000 \\
\end{pmatrix} $


If we move to numpy, we can see another illustration of this point.

```python
import numpy as np
```

```python
a = np.array([800, 1500, 2000, 3500, 4000])
```

```python
a*2
```

One of the main takeaways from the above, is that a number multiplied by a vector, returns to us another vector.  And it's calculated out multiplying each entry in a vector by that number.

So multiplying a vector by a scalar just means stretching or shrinking the vector.  To accomplish this, we multiply each component of the vector by the same number.

* Vector addition

Vector matrix addition is applied entrywise over the matrices.  So we just add up entries with the same index.

```python
a = np.array([1200, 300, -400, 600, 200])

b = np.array([800, 1500, 2000, 3500, 4000])

a + b 
```

* Matrix addition

```python
A = np.array([
    [800, 220],
    [1500, 440],
    [2000, 890],
    [3500, 250],
    [4000, 700],
])

B = np.array([
    [300, 330],
    [1200, 110],
    [700, 900],
    [890, 120],
    [1410, 100],
])

A + B
```

As you can see, matrix addition occurs in the same way.  Entries with the same index are added together.  And the same rules apply for vector and matrix subtraction.

```python
a = np.array([1200, 300, -400, 600, 200])

b = np.array([800, 1500, 2000, 3500, 4000])

a - b
```

Here, the operation just matches up elements with identical indices and then subtracts.

### Matrix Operations and Shape

* But beware

To add or subtract vectors from one another, they must have the same dimensions. 

```python
a = np.array([1200, 300, -400, 600, 200])

c = np.array([200, 300, 700, 4000])
a + c
```

```python
a.shape
# (5,)
c.shape 
# (4,)
```

And, of course, matrices must also have identical shape, to add or subtract them.

```python
A = np.array([
    [800, 220],
    [1500, 440],
    [2000, 890],
    [3500, 250],
    [4000, 700],
])

C = np.array([[ 400, 1500,  300,  1200, 670],
       [ 760,  450,  210,  450,  960]])

A.shape
# (5, 2)

C.shape

# (2, 5)
```

Above, 
* A has five rows and two columns
* C has two rows and five columns.  

Will it add?

```python
A + C
```

No it won't.  

But we can fix it.  $ A \in \mathbb{R}^{5,2} $ and $ C \in \mathbb{R}^{2, 5} $.  One solution is to flip the rows and columns of the matrix C.  That is to transpose C to make it a 5 by 2 matrix, and then add C + A.  

Let's see it.

```python
C = np.array([[ 400, 1500,  300,  1200, 670],
       [ 760,  450,  210,  450,  960]])
```

```python
C.T
```

We just transposed the matrix.  This changes the entries in the first row, to entries in the first column.  Entries in the second row are now the entries in the second column.  And so on.  

Now notice that A and $C^{T}$, C transposed, have the same shape.

```python
A.shape
```

```python
C.T.shape
```

So we can add these matrices up.

```python
A
```

```python
C.T
```

```python
A + C.T
```
