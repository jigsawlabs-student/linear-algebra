# Applied PCA

### Introduction

Principal component analysis is another mechanism of feature reduction, often used with high dimensional matrices, like with images or text.  It allows us to see patterns in the data, that we might not otherwise.

Here, we'll what are principal component analysis is, and how we can use principal component analysis to reduce our features.

### Creating Data

Let's say that we are plotting features regarding movie revenue and we plot number of spoken words in a movie and number of songs in a movie.  We notice something like the following:  

```python
import numpy as np
np.random.seed(0)

mean = np.array([100, 20])
cov = np.array([[1, 1],
               [2, 3]])
```

```python
X = np.random.multivariate_normal(mean, cov, 300)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 6))
plt.scatter(X[:, 0],
            X[:, 1],
            s=40, c = 'r')
plt.title('Number of songs vs spoken words')
ax = plt.gca()
ax.set_ylabel('num songs')
ax.set_xlabel('num words')
```

We can see that we have two highly correlated features.  And that, perhaps we can represent these as a new feature.  Perhaps, amount of content in the movie, or perhaps we have a signal of movie length.  

So, we would like to reduce these two features into a single feature.  To perform this, we can use PCA.

### Applying PCA

Let's get a sense of how to use principal component analysis through scikit learn.

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)
```

We employ PCA by initializing an instance of PCA and then calling `fit`, providing in our feature data.  After doing that, our pca instance returns to us a set of components.

```python
pca.components_
```

Let's take a look at what these components identify for us.

> We need to center our data before re-plotting it.

```python
X_centered = X - X.mean(axis = 0)
```

```python
plt.figure(figsize = (6, 6))
plt.scatter(X_centered[:, 0],
            X_centered[:, 1],
            s=40, c = 'r')

for l, v in zip(pca.explained_variance_ratio_, pca.components_):
    d = 5* np.sqrt(l) * v
    plt.plot([0, d[0]], [0, d[1]], '-k', lw=3)
```

We can see that the principal components identify the directions in which there is movement in the data.  Essentially, PCA asks, if we were to reduce our data from two dimensions to one dimension what is the vector that retain the movement in the data.  The vector that performs this, is the vector along which there is the most amount of variation in the data. 

### Explain 90% of variance

The reason why this is important, is because the next step is for PCA to simply reduce our two dimensional points to a single number which represents where the point lies along the first principal component.  And the point is that, even though we are reducing the number of dimensions from two to one, we are still capturing the movement in the data. 

Let's take a look at how PCA transforms our data.  We'll start with our original data X, and then reduce it by initializing PCA, and calling `fit_transform`.

Our data starts off in two dimensions.

```python
X[:5]
```

```python
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)
```

But by identifying the vector with the largest amount of variance, PCA expresses this data in terms of a single dimension.

```python
X_reduced[:5]
```

We can get a sense of how this is accomplished by transforming and then `inverse_transforming` the data.

```python
X_new = pca.inverse_transform(X_reduced[:20])

plt.figure(figsize=(5,5))
plt.plot(X[:20, 0], X[:20, 1], 'or', alpha=0.3)
plt.plot(X_new[:20, 0], X_new[:20, 1], 'or', alpha=0.8)
plt.axis('equal')
plt.title('Projection of sample onto the 1st principal component')
plt.show()
```

From the picture above, we can get a sense of what PCA is doing.  It identifies the vector of greatest variance, and then projects the data onto this vector to maintain the variation in the data. By doing this, PCA is able to capture a lot of the variation in the data while reducing the number of features.

### PCA what is it good for?

Principal component analysis has a few different uses.  
* Data compression.  If the same information can be represented with fewer features, then we can save space in storing that data.  This is true whether the application is machine learning, or simply saving a sound file in less space.  By representing the data in a format that requires less data, that data can be compressed.

* Training Speed.  When we have lots of features, like a text, sound or image file training speed can become an issue.  By reducing the number of features for each observation, we can speed up training time of the machine learning algorithm.

* Data visualization.  When working with datasets that have large number of dimensions, it can become difficult to see how the data clusters together.  PCA can be used to reduce the number of dimensions so that data can be plotted, yet still capture differences in the observations.
